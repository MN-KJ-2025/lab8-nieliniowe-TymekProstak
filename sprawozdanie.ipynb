### **Zadanie 5.**

Dla funkcji $f(x)$ zdefiniowanej w sekcji **Cel zajęć** znajdź miejsca zerowe przy użyciu:

- funkcji [`sp.optimize.root()`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.root.html).
- funkcji [`sp.optimize.fsolve()`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.fsolve.html).

```python
# ===== Twoja implementacja tutaj =====
# korzystamy z tego że zgadujemy że rozwiązania jest  w przedziale [0,1] korzystając z wykresu f(x)
x1_start_guess = -0.5
x2_start_guess = 1
x1_root = sp.optimize.root(main.func, x1_start_guess).x
x2_root = sp.optimize.root(main.func, x2_start_guess).x
x1_fsolve = sp.optimize.fsolve(main.func, x1_start_guess)
x2_fsolve = sp.optimize.fsolve(main.func, x2_start_guess)
print('Miejsca zerowe znalezione przy pomocy funkcji roots: \n{0}, {1}\n'.format(x1_root, x2_root))
print('Miejsca zerowe znalezione przy pomocy funkcji fsolve: \n{0}, {1}\n'.format(x1_fsolve, x2_fsolve))
```

### **Zadanie 6.**

Korzystając z przedziałów wyznaczonych w **Zadaniu 1.** znajdź miejsca zerowe funkcji $f(x)$ zdefiniowanej w sekcji **Cel zajęć**, przy użyciu:

- metody bisekcji,
- metody siecznych,
- metody Newtona,

z tolerancją równą $1\mathrm{e}{-10}$.

Zbadaj dokładność (względem rozwiązania z **Zadania 5.**) i czas obliczeń metod w zależności od liczby iteracji. Wyniki przedstaw na wykresach. 

```python
# ===== Twoja implementacja tutaj =====


import scipy.optimize
import timeit
import main

x1_range = [-0.5, 0.5]
x2_range = [0.5, 2]
epsilon = 1e-10
iteration = 100

x1_root = scipy.optimize.root(main.func, x1_range[0]).x
x2_root = scipy.optimize.root(main.func, x2_range[1]).x


start = timeit.default_timer()
x1_bisection = main.bisection(x1_range[0], x1_range[1], main.func, epsilon, iteration)
end = timeit.default_timer()
time_x1_bis = end - start

start = timeit.default_timer()
x2_bisection = main.bisection(x2_range[0], x2_range[1], main.func, epsilon, iteration)
end = timeit.default_timer()
time_x2_bis = end - start


start = timeit.default_timer()
x1_secant = main.secant(x1_range[0], x1_range[1], main.func, epsilon, iteration)
end = timeit.default_timer()
time_x1_sec = end - start

start = timeit.default_timer()
x2_secant = main.secant(x2_range[0], x2_range[1], main.func, epsilon, iteration)
end = timeit.default_timer()
time_x2_sec = end - start


start = timeit.default_timer()
x1_newton = main.newton(main.func, main.dfunc, main.ddfunc, x1_range[0], x1_range[1], epsilon, iteration)
end = timeit.default_timer()
time_x1_new = end - start

start = timeit.default_timer()
x2_newton = main.newton(main.func, main.dfunc, main.ddfunc, x2_range[0], x2_range[1], epsilon, iteration)
end = timeit.default_timer()
time_x2_new = end - start



print('-==--=-=-=-==-=-=--=-==--==-=--=-==-=--==-')
print('Rozwiązanie uzyskane z funkcji roots: \n')
print('Pierwsze miejsce zerowe: {0},\nDrugie miejsce zerowe: {1}.\n'.format(x1_root[0], x2_root[0]))


print('-==--=-=-=-==-=-=--=-==--==-=--=-==-=--==-')
print('Rozwiązanie uzyskane metodą siecznych: \n')
print('Pierwsze miejsce zerowe: {0}\nUzyskane w {1} iteracjach.\nCzas obliczeń: {2:.8f} s\n'.format(x1_secant[0], x1_secant[1], time_x1_sec))
print('Drugie miejsce zerowe: {0}\nUzyskane w {1} iteracjach.\nCzas obliczeń: {2:.8f} s\n'.format(x2_secant[0], x2_secant[1], time_x2_sec))
print('Błąd wyliczeń względem metody root wynosi:\n- dla x1: {0},\n- dla x2: {1}\n'.format(abs(x1_root[0]-x1_secant[0]), abs(x2_root[0]-x2_secant[0])))

print('-==--=-=-=-==-=-=--=-==--==-=--=-==-=--==-')
print('Rozwiązanie uzyskane metodą Newtona: \n')
print('Pierwsze miejsce zerowe: {0}\nUzyskane w {1} iteracjach.\nCzas obliczeń: {2:.8f} s\n'.format(x1_newton[0], x1_newton[1], time_x1_new))
print('Drugie miejsce zerowe: {0}\nUzyskane w {1} iteracjach.\nCzas obliczeń: {2:.8f} s\n'.format(x2_newton[0], x2_newton[1], time_x2_new))
print('Błąd wyliczeń względem metody root wynosi:\n- dla x1: {0},\n- dla x2: {1}\n'.format(abs(x1_root[0]-x1_newton[0]), abs(x2_root[0]-x2_newton[0])))
print('-==--=-=-=-==-=-=--=-==--==-=--=-==-=--==-')

print('-==--=-=-=-==-=-=--=-==--==-=--=-==-=--==-')
print('Rozwiązanie uzyskane metodą bisekcji: \n')
print('Pierwsze miejsce zerowe: {0}\nUzyskane w {1} iteracjach.\nCzas obliczeń: {2:.8f} s\n'.format(x1_bisection[0], x1_bisection[1], time_x1_bis))
print('Drugie miejsce zerowe: {0}\nUzyskane w {1} iteracjach.\nCzas obliczeń: {2:.8f} s\n'.format(x2_bisection[0], x2_bisection[1], time_x2_bis))
print('Błąd wyliczeń względem metody root wynosi:\n- dla x1: {0},\n- dla x2: {1}\n'.format(abs(x1_root[0]-x1_bisection[0]), abs(x2_root[0]-x2_bisection[0])))
print('-==--=-=-=-==-=-=--=-==--==-=--=-==-=--==-')
```

Metoda Newtona charakterzuję się najwieskzym błedem, metoda bisekcji najdokładniejszym rozwiązaniem, ale jednocześnie najwiekszą ilością iteracji. Metoda Newtona
